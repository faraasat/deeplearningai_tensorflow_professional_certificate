## 2) Importing Required Libraries


```python
import tensorflow as tf
from tensorflow import keras as ks
import numpy as np
```

## 3) Creating a Model


```python
def house_model():
    # Defining x and y
    xs = np.array([1, 2, 3, 4, 5, 6], dtype="float")
    ys = np.array([1.0, 1.5, 2.0, 2.5, 3.00, 3.50], dtype="float")

    # Creating a Sequential Model from keras
    # Using a Dense Layer with 1 unit and input shape of 1
    # units represents the dimensionality of the output space
    # input_shape is the dimensions of input data
    model = ks.Sequential([ks.layers.Dense(units=1, input_shape=[1])])

    # Compiling model and optimizer is set to the Stochastic Gradient Descent
    # and Mean Squared Error is used as the loss function
    model.compile(optimizer="sgd", loss="mean_squared_error")

    # Fitting the data and training the model for 1000 epochs
    model.fit(xs, ys, epochs=1000)

    # return the model
    return model
```

## 4) Calling the Model


```python
model = house_model()
```

## 5) Getting Predictions From the Model


```python
new_x = [7.0, 8.0, 8.5, 9.0]
prediction = model.predict(new_x)
print(prediction)
```
